## Baseline Models

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, StratifiedKFold

def cross_val_scores_by_clf(clfs, X, y, cv):
    '''
    Creates a dictionary with clf names as keys and list of test scores for each fold of the cross validation

    Parameters
    ----------
    clfs : list of tuples with classifier name and instantiated classifier
    
    cv : cross validation to use for cross_val_score
    '''
    score_by_clf = {}
    
    for name, clf in clfs:
        score_by_clf[name] = cross_val_score(clf, X, y, cv=cv)

    return score_by_clf

def append_mean_std_to_df(df, identifier=''):
    df['mean' + identifier] = df.mean(axis=1)
    df['std' + identifier] = df.std(axis=1)
    return df
    
def score_by_clf_to_df(score_by_clf, i=''):
    '''
    Creates a dataframe from a dictionary of classifier names and scores, 
    appending two new columns of computed mean and std of scores
    
    Parameters:
    score_by_clf : dictionary with classifier names as keys and a list of scores as values
    i : optional, denotes the iteration of cv, used in naming when appending new mean and std columns
    '''
    score_by_clf_df = pd.DataFrame.from_dict(score_by_clf, orient='index')
    score_by_clf_df = append_mean_std_to_df(score_by_clf_df, i)

    return score_by_clf_df
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True)
clfs = [
    ('knn', KNeighborsClassifier()),
    ('logreg', LogisticRegression()),
    ('svm', SVC()),
    ('nb', GaussianNB()),
    ('dtree', DecisionTreeClassifier()),
    ('rforest', RandomForestClassifier())
]

X = iris.drop('Species', axis=1)
y = iris['Species']
```

```python
# one run of cross validation
score_by_clf = cross_val_scores_by_clf(clfs, X, y, kfold)
score_by_clf_df = score_by_clf_to_df(score_by_clf)
score_by_clf_df
```

```python
# running cross validation a few times to check the stability of the mean
scores = []
for i in range(10):
    score_by_clf = cross_val_scores_by_clf(clfs, X, y, kfold)
    score_by_clf_df = score_by_clf_to_df(score_by_clf, '_cv' + str(i))
    scores.append(score_by_clf_df)
```

```python
means = []
stds = []
for i, df in enumerate(scores):
    means.append(df['mean_cv' + str(i)])
    stds.append(df['std_cv' + str(i)])
    
means_df = pd.DataFrame(means).T
means_df.plot(figsize=(10,10))
plt.show()
```

```python
mean_of_means = np.mean(means_df, axis=1)
std_of_means = np.std(means_df, axis=1)

print('Mean\n', mean_of_means.sort_values(ascending=False), '\n')
print('Std\n', std_of_means.sort_values())
```

After iterating 5-fold cross validation 10 times, we find that all models perform farily well out-of-the-box with test scores above 0.94 and small variance (all on the order of at most 10^-2). SVM does particularly well with consistently the highest average test scores. The ranking of models based on average mean stay consistent. The variance, on the otherhand, although small, isn't stable enough to rank the different models.

## Model tuning

Use hyperparameter tuning and feature selection techniques to try to boost accuracy for some of these models. I've chosen, a decision tree, svm, and knn to tune.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest
```

```python
seed = 42
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
```

### Decision Tree

```python
pipeline = Pipeline([
        ('reduce_dim', PCA()),
        ('dtree', DecisionTreeClassifier(random_state=seed))
    ])

param_grid = [
    {
        'reduce_dim__n_components' : range(1,5),
        'dtree__max_depth' : range(1,10)
    },
    {
        'reduce_dim': [SelectKBest()],
        'reduce_dim__k' : range(1,4),
        'dtree__max_depth' : range(1,10)
    }

]
estimator = GridSearchCV(pipeline, param_grid=param_grid, cv=kfold)
estimator.fit(X, y)
print('Best params:', estimator.best_params_)
print('Best score:', estimator.best_score_)
```

We use a pipeline with PCA for feature selection and a Decision Tree Classifier. 
Tuning hyperparameters using an exhaustive grid search yields optimal parameters:
- PCA with n_components = 2 and
- Decision tree with max_depth = 5

```python
from sklearn.tree import export_graphviz
import graphviz

dtree = estimator.best_estimator_.named_steps['dtree']
data = export_graphviz(dtree, out_file=None,
                           feature_names=['pca_cmp1', 'pca_cmp2'],
                           class_names=dtree.classes_.astype('<U10'),
                           filled=True, rounded=True,
                           special_characters=True)
graph = graphviz.Source(data)
graph
```

```python
dtree.feature_importances_
```

We can create a graphical visualization of the decision tree using graphviz. As the features used as inputs to the decision tree are the outputs of PCA, the tree isn't very interpretable beyond speaking in terms of PCA components.

### KNearestNeighbors

```python
pipe = Pipeline([('knn', KNeighborsClassifier())])
K_NEIGHBORS = range(1, 50)
param_grid = [{'knn__n_neighbors': K_NEIGHBORS}]
estimator = GridSearchCV(pipe, param_grid=param_grid, cv=kfold)
estimator.fit(X, y)
estimator.best_estimator_
print(estimator.best_params_)
print(estimator.best_score_)
```

```python
mean_test_score = estimator.cv_results_['mean_test_score']
mean_train_score = estimator.cv_results_['mean_train_score']

plt.plot(K_NEIGHBORS, mean_test_score)
plt.plot(K_NEIGHBORS, mean_train_score)
```
